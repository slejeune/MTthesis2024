```python
# Imports
import matplotlib.pyplot as plt
```

```python
def load_data(lang1:str, lang2:str):
    with open('/Users/Suzenator/Documents/Uni/M4/MThesis/Data/Non-English-raw/'+lang1+'-'+lang2+'/NLLB.'+lang1+'-'+lang2+'.'+lang1) as f:
        lang1_lang2_pre = f.readlines()
    
    with open('/Users/Suzenator/Documents/Uni/M4/MThesis/Data/Non-English-raw/'+lang1+'-'+lang2+'/NLLB.'+lang1+'-'+lang2+'.'+lang2) as f:
        lang2_lang1_pre = f.readlines()

    with open('/Users/Suzenator/Documents/Uni/M4/MThesis/Data/Non-English-preprocessed/'+lang1+'-'+lang2+'/NLLB.'+lang1+'-'+lang2+'.'+lang1) as f:
        lang1_lang2_post = f.readlines()
    
    with open('/Users/Suzenator/Documents/Uni/M4/MThesis/Data/Non-English-preprocessed/'+lang1+'-'+lang2+'/NLLB.'+lang1+'-'+lang2+'.'+lang2) as f:
        lang2_lang1_post = f.readlines()

    return lang1_lang2_pre, lang2_lang1_pre, lang1_lang2_post, lang2_lang1_post
```

```python
def plot_clen(lang1_lang2_pre, lang2_lang1_pre, lang1_lang2_post, lang2_lang1_post):
    clen12_pre = sorted([len(sent) for sent in lang1_lang2_pre])
    clen21_pre = sorted([len(sent) for sent in lang2_lang1_pre])

    clen12_post = sorted([len(sent) for sent in lang1_lang2_post])
    clen21_post = sorted([len(sent) for sent in lang2_lang1_post])

    fig, ax = plt.subplots(2,2,figsize=(7,6))
    
    # Before
    ax[0,0].plot(clen12_pre)
    ax[0,0].set_title(lang1+"-"+lang2+"\n(raw)",fontsize=10)
    ax[0,0].set_xlabel("Sentences (n)")
    ax[0,0].grid(axis='y', linestyle='--', alpha=0.7)
    ax[0,0].set_ylabel("Characters (n)")
    ax[0,1].plot(clen21_pre)
    ax[0,1].set_title(lang2+"-"+lang1+"\n(raw)",fontsize=10)
    ax[0,1].set_xlabel("Sentences (n)")
    ax[0,1].set_ylabel("Characters (n)")
    ax[0,1].grid(axis='y', linestyle='--', alpha=0.7)
    
    # After
    ax[1,0].plot(clen12_post, c='orange')
    ax[1,0].set_title(lang1+"-"+lang2+"\n(preprocessed)",fontsize=10)
    ax[1,0].set_xlabel("Sentences (n)")
    ax[1,0].set_ylabel("Characters (n)")
    ax[1,0].grid(axis='y', linestyle='--', alpha=0.7)
    ax[1,1].plot(clen21_post,c='orange')
    ax[1,1].set_title(lang2+"-"+lang1+"\n(preprocessed)",fontsize=10)
    ax[1,1].set_xlabel("Sentences (n)")
    ax[1,1].set_ylabel("Characters (n)")
    ax[1,1].grid(axis='y', linestyle='--', alpha=0.7)

#     fig.suptitle("Sentence length (in characters) before and after preprocessing")
    fig.tight_layout()
    plt.show()
```

```python
dirs = ['fr-ja','ja-tl','ja-nl','fr-tl','fr-nl','tl-nl', # high -> high
       'ja-ne','fr-ig','tl-cy','nl-li', # high -> low
       'ne-ig','ne-cy','ne-li','ig-cy','ig-li','cy-li'] # low -> low
```

```python
for dir in dirs:
    lang1, lang2 = dir.split('-')
    lang1_lang2_pre, lang2_lang1_pre, lang1_lang2_post, lang2_lang1_post = load_data(lang1,lang2)
    plot_clen(lang1_lang2_pre, lang2_lang1_pre, lang1_lang2_post, lang2_lang1_post)
```
